###  Creazione  object vessel "Vs" : Lista di tutti gli ais data suddivisa per vascello

```python
SINGLE_VESSEL_id = (pd.unique(ais_data['MMSI']))
Vs={}
for i in tqdm(range(0,len(SINGLE_VESSEL_id))):
    Vs[i]=ais_data[ais_data['MMSI']==SINGLE_VESSEL_id[i]]
    Vs[i].reset_index(inplace = True, drop = True)
```

aggiungiamo il campo  "status" e "avg_speed" 

status= sailing/stationary in base alla velocità calcolata utilizzando le info di lat e long


```python

for i in tqdm(range(0,len(Vs))):
    Vs[i]['status'] = 0 
    Vs[i]['avg_speed'] = 0 


newVs = {}
k=0
# eliminate the unrequired element --- teniamo soltanto vascelli che hanno almeno più di 50 istanze
for key, value in Vs.items():
    if len(value) >=50 :
        newVs[k] = value
        k=k+1

Vs=newVs
print('lunghezza vettore VS ='+ str(len(Vs)))
```

 calcoliamo la velocità

```python
from datetime import datetime


d2m = 4E+7/360;                                                    

hypot = np.vectorize(math.hypot)

for i in tqdm(range(0,len(Vs))):
    
    temp=Vs[i]

    temp['BaseDateTime'] = pd.to_datetime(temp['BaseDateTime'] , format='%d/%m/%Y %H:%M')
    d=(temp['BaseDateTime'] - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s')
    #d=(temp['t'].to_numpy())/1000
    
    Tsec=d-d[0]

    deltaT=np.diff(Tsec)

    deltaT=np.append(0, deltaT)
    dLat = np.diff(temp['LAT'])
    dLon = np.diff(temp['LON'])
    DistDeg = hypot(dLat, dLon)                       
    DistMtr = DistDeg * d2m                                         
    dTime = np.diff(Tsec)                                               
    Velocity = DistMtr / dTime
    Vs[i]['avg_speed']=np.append(0, Velocity)
    Vs[i]['status']=Vs[i]['avg_speed'] ==0
    Vs[i]['status'] = [str(x) for x in  Vs[i]['status']]
    Vs[i]['status'][Vs[i]['status']=='True']='stationary'
    Vs[i]['status'][Vs[i]['status']=='False']='sailing'

    Vs[i] = Vs[i].fillna(0)

```

### Creazione oggetto WPs (punti stazionari)
Wp unione dei PO(punti stazionari), EN (punti d'ingresso) e EX(punti d'uscita)


```python
df = pd.concat(Vs)
idx = df['status'] == 'stationary'
PO =  df[idx]

#eliminiamo i duplicati
#PO=PO.drop_duplicates(subset=['LAT', 'LON'], keep='first')


ENs=pd.DataFrame()
#prendiamo i primi campi di ogni vascello
for i in tqdm(range(0,len(Vs))):
    ENs[i]=Vs[i].iloc[0]
ENs=ENs.T 
#ENs=ENs.drop_duplicates(subset=['LAT', 'LON'], keep='first')


EXs=pd.DataFrame()
#prendiamo i primi campi di ogni vascello
for i in tqdm(range(0,len(Vs))):
    EXs[i]=Vs[i].iloc[-1]

EXs=EXs.T 
#EXs=EXs.drop_duplicates(subset=['LAT', 'LON'], keep='first')
```

```python
#unione dataset
WP=pd.concat([PO,ENs,EXs])
WP=WP.drop_duplicates(subset=['LAT', 'LON'], keep='first')
```

```python
#estriamio soltanto lat e lon per visualizzare i cluster sulla mappa
subsetWP=WP[['LAT', 'LON']]

#riscaliamo per DBSCAN
scaler = StandardScaler()
scaled_array = scaler.fit_transform(subsetWP)
```

```python
#applichiamo DBSCAN all'array riscalato
db = DBSCAN(eps=0.2, min_samples=20).fit(scaled_array) ## impostare gli iperparametri eps e min_samples, dipende dal dataset. l'obiettivo dovrebbe essere trovare i cluster nelle zone portuali

#aggiungiamo le labels del cluster d'appartenza al set Wp
labels = db.labels_
subsetWP["LABEL"] = labels
subsetWP.hvplot.scatter(x = "LON", y = "LAT", by='LABEL')
```

```python
#creazione di poligoni convessi come cluster
from shapely.geometry import Polygon

subsetWP['LABEL'].nunique()
WP_hull=[]
WP_I=[]
##rimuove label = -1, corrisponde all'anomalie/ appartenenza a nessun cluster
WP_lab=pd.unique(subsetWP['LABEL'])
WP_lab = np.delete(WP_lab, np.where(WP_lab==-1))
for i in tqdm(range(0,len(WP_lab))):
          WP_I.append(subsetWP[['LON','LAT']][subsetWP["LABEL"]==WP_lab[i]].to_numpy())
          p = Polygon((WP_I[i]))

          WP_hull.append(p.convex_hull)
```

```python
Route_object=pd.DataFrame()
# ogni volta che nel dataframe la nave passa per più Wp, la funzione
# estre la porzione di dati che va da un Wp_X1 a Wp_X2, tralasciando il resto

for index_possible_route in range(0,len(possible_vessel_route)):
    #estrazione rotte
    temp=possible_vessel_route[index_possible_route]
    start_r=[]
    end_r= []
    i=0
    index_label=0
    array_label=[]
    while i < len(temp):
        if temp['waypoints'][i]=='-1':
            if i==0:
                start_temp=0
                start_wp=0
            else:
                start_temp=i
                start_wp=i-1

            while temp['waypoints'][i]=='-1' and i < len(temp)-1:
                i=i+1

            end_temp=i-1
            end_wp=i
            

            if temp['waypoints'][start_wp]!=temp['waypoints'][end_wp] and  temp['waypoints'][start_wp]!='-1' and  temp['waypoints'][end_wp]!='-1' :
                start_r.append(start_temp)
                end_r.append(end_temp)
                array_label.append(temp['waypoints'][start_wp]+'-'+temp['waypoints'][end_wp])


        i=i+1

    for i in range(0,len(array_label)):
        route_temp=(possible_vessel_route[index_possible_route].iloc[ start_r[i]-1 :end_r[i]+1 ,:])

        route_temp['Route']=array_label[i]
        Route_object = pd.concat([Route_object, route_temp.reset_index()])
```
